```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file=pd.read_csv('ODI_data.csv')
```

```python
file
```

```python
sachin=file[file['Innings Player']=='SR Tendulkar']
```

```python
sachin
```

```python
sachin_runs=sachin[['Innings Runs Scored','Opposition']]
```

```python
sachin_runs
```

```python
sachin_runs=sachin_runs.dropna()
```

```python
sachin_runs=sachin_runs[~sachin_runs['Innings Runs Scored'].isin(['TDNB','DNB'])]
```

```python
sachin_runs
```

```python
sachin_notout=sachin_runs[sachin_runs['Innings Runs Scored'].str.contains('*','',regex=False)]
```

```python
sachin_notout
```

```python
sachin_notout.loc[:,'Innings Runs Scored']=sachin_notout['Innings Runs Scored'].str.replace('*','',regex=False)
```

```python
sachin_notout.loc[:,'Innings Runs Scored']=sachin_notout['Innings Runs Scored'].astype(int)
```

```python
sachin_notout=sachin_notout['Innings Runs Scored'].count()
```

```python
sachin_notout
```

```python
sachin_runs.loc[:,'Innings Runs Scored']=sachin_runs['Innings Runs Scored'].str.replace('*','',regex=False)
sachin_runs.loc[:,'Innings Runs Scored']=sachin_runs['Innings Runs Scored'].astype(int)
```

```python
sachin_total_runs=sachin_runs['Innings Runs Scored'].sum()
sachin_total_runs

sachin_runs['Innings Number']=np.arange(1,len(sachin_runs)+1)
sachin_runs = sachin_runs[['Innings Number'] + [col for col in sachin_runs.columns if col != 'Innings Number']]
sachin_runs
```

```python
sachin_total_innings=sachin_runs['Innings Number'].max()
sachin_total_innings
```

```python
sachin_batting_average=sachin_total_runs/(sachin_total_innings-sachin_notout)
sachin_batting_average
```

```python
virat=file[file['Innings Player']=='V Kohli']
virat
```

```python
virat_runs=virat[['Innings Runs Scored','Opposition']]
virat_runs=virat_runs.dropna()
virat_runs=virat_runs[~virat_runs['Innings Runs Scored'].isin(['TDNB','DNB'])]
virat_notout=virat_runs[virat_runs['Innings Runs Scored'].str.contains('*','',regex=False)]
virat_notout.loc[:,'Innings Runs Scored']=virat_notout['Innings Runs Scored'].str.replace('*','',regex=False)
virat_notout.loc[:,'Innings Runs Scored']=virat_notout['Innings Runs Scored'].astype(int)
virat_notout=virat_notout['Innings Runs Scored'].count()
virat_runs.loc[:,'Innings Runs Scored']=virat_runs['Innings Runs Scored'].str.replace('*','',regex=False)
virat_runs.loc[:,'Innings Runs Scored']=virat_runs['Innings Runs Scored'].astype(int)
virat_runs['Innings Number']=np.arange(1,len(virat_runs)+1)
virat_runs = virat_runs[['Innings Number'] + [col for col in sachin_runs.columns if col != 'Innings Number']]
virat_total_innings=virat_runs['Innings Number'].max()
virat_total_runs=virat_runs['Innings Runs Scored'].sum()
virat_batting_average=virat_total_runs/(virat_total_innings-virat_notout)
virat_batting_average
```

```python
plt.figure(figsize=(14,8))
plt.plot(sachin_runs['Innings Number'],sachin_runs['Innings Runs Scored'])
plt.title(" Sachin Tendulkar ODI Batting Stats:Innings Runs vs Innings Number")
plt.xlabel("Innings Number")
plt.ylabel("Innings Runs Scored")
plt.show()

plt.figure(figsize=(14,8))
plt.plot(virat_runs['Innings Number'],virat_runs['Innings Runs Scored'])
plt.title("Virat Kohli ODI Batting Stats:Innings Runs vs Innings Number")
plt.xlabel("Innings Number")
plt.ylabel("Innings Runs Scored")
plt.show()
```
